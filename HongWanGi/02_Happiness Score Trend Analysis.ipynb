# Happiness Score Trend Analysis
이 노트북은 주요 국가(대한민국, 일본, 미국, 영국, 독일)의 연도별 행복 지수 변화를 분석하고 시각화합니다.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Load Data
통합 데이터(`happiness_2015_2024.csv`)를 불러옵니다.

```python
happiness_df = pd.read_csv('../data/happiness_2015_2024.csv')
```

## 2. Filter Target Countries
분석 대상이 되는 주요 6개국 데이터만 추출합니다.( 한국, 미국, 영국, 독일, 일본, 중국 )

```python
target_countries = ["South Korea", "Japan", "United States", "United Kingdom", "Germany", "China"]
major_countries_df = happiness_df[happiness_df["country"].isin(target_countries)].copy()
```

## 3. Visualization
연도별 행복 지수 추이를 선 그래프로 그리고, 대한민국의 추이를 강조합니다.

```python
custom_palette = {
    "South Korea": "#FF0000",
    "Japan": "#9467bd",
    "United States": "#1f77b4",
    "United Kingdom": "#2ca02c",
    "Germany": "#ff7f0e",
    "China": "#8c564b"
}

fig, ax = plt.subplots(figsize=(12, 7))

sns.lineplot(
    data=major_countries_df,
    x="year",
    y="score",
    hue="country",
    palette=custom_palette,
    marker="o",
    linewidth=2.5,
    markersize=8,
    ax=ax
)

for line in ax.lines:
    if line.get_color() == "#FF0000":
        line.set_linewidth(4.0)
        line.set_markersize(10)
        line.set_zorder(5)
    else:
        line.set_linewidth(2.0)
        line.set_alpha(0.8)

ax.set_title("Happiness Score Trends of Major Countries (2015-2024)", pad=15)
ax.set_xlabel("Year")
ax.set_ylabel("Happiness Score")
ax.set_xticks(range(2015, 2025))
ax.grid(True, linestyle="--", alpha=0.7)

plt.tight_layout()
plt.show()
```
